# Анализ признаков преступников и потерпевших в Российской Федерации

## Описание исследования

Правоохранительные органы Российской Федерации осуществляют сбор статистики по лицам совершившими преступления, а так же лицам, которые признаны потерпевшими в результате противоправной деятельности.

## Цель исследования

На основе имеющихся данных, установить характерные признаки лиц совершивших преступления, и лиц признаных потерпевшими

## Задачи исследования

- Исследовать половозрастной портрет преступников и потерпевших в Российской Федерации
- Исследовать динамику изменения значений половозрастных признаков по годам
- Выявить регионы Российской Федерации, которые отличаются от общих тенденций по половозрастным признакам преступников и потерпевших

## Исходные данные

__gender_age_crimes__ - количество выявленных преступников в период 2014-2021 гг, их половозрастные признаки:

- `region` — субъект Российской Федерации;
- `offender_feature` - признак преступника:
    - выявлено лиц, совершивших преступления
    - выявлено лиц, совершивших преступления, в том числе женщин
    - выявлено лиц, совершивших преступления, в том числе несовершеннолетних
- `total` - количество преступников
- `year` - год
---

__deviations_influence_crimes__ - количество выявленных преступников в период 2015-2021 гг, в состоянии алкогольного, наркотического, токсического опьянения, а также лиц повторно совершавших преступления:

- `region` — субъект Российской Федерации;
- `offender_feature` - признак преступника:
    - количество расследованных преступлений, совершенных лицами, ранее совершавшими преступления
    - количество расследованных преступлений, совершенных в состоянии алкогольного опьянения
    - количество расследованных преступлений, совершенных в состоянии наркотического опьянения
    - количество расследованных преступлений, совершенных в состоянии токсического опьянения
- `total` - количество преступников
- `year` - год
---

__victims__ - количество потерпевших в период 2015-2021 гг, их юридический статус и половозрастные признаки

- `region` — субъект Российской Федерации;
- `victim_feature` - признак потерпевшего:
    - количество потерпевших
    - количество потерпевших, в том числе юридических лиц
    - количество потерпевших, в том числе физических лиц
    - количество потерпевших, в том числе погибших
    - количество потерпевших, в том числе которым причинен тяжкий вред здоровью
    - количество потерпевших, в том числе несовершеннолетних
    - количество потерпевших, в том числе женщин
- `total` - количество потерпевших
- `year` - год
---

Источник информации: [Открытые данные МВД Российской Федерации](https://xn--b1aew.xn--p1ai/opendata/)

## Исследовательский анализ

### Исследуем признаки преступников для всей России за весь имеющийся период

```python
offender_feature_grouped = (
    gender_age_crimes_df.query('region == "Всего по России "')
                        .groupby('sex_age_feature')
                        .agg({'total': 'sum'})
                        .reset_index()
                        .sort_values(by='total', ascending=False)
)
offender_feature_grouped
```

Получили таблицу с количеством выявленных преступников по признакам. Теперь добавим в таблицу наблюдение, со значением количества преступлений совершенных мужчинами старше 18 (для наглядности)

```python
man_crimes_cnt = np.sum(
    [offender_feature_grouped['total'][0], 
     *offender_feature_grouped['total'][1:] * -1]
)

man_crimes = pd.DataFrame(data={'sex_age_feature': ['man'], 'total': [man_crimes_cnt]})
offender_feature_grouped = pd.concat([offender_feature_grouped, man_crimes])
offender_feature_grouped
```

Посчитаем доли каждого признака, от общего количества выявленных преступников

```python
offender_feature_grouped['ratio'] = (
    offender_feature_grouped['total'].div(offender_feature_grouped['total'].values[0])
)
```

```python
offender_feature_grouped
```

Выводы:
- В период с 2014-2021 годы выявлено 4 257 144 преступников;
- Доля женщин среди выявленных преступников - 16%;
- Доля несовершеннолетних - 4%;

### Исследуем признаки потерпевших для всей России без учета года

```python
victim_feature_grouped = (
    victims_df.query('region == "Всего по России "')
                        .groupby('victim_feature')
                        .agg({'total': 'sum'})
                        .sort_values(by='total', ascending=False)
                        .reset_index()
)
victim_feature_grouped
```

Получили таблицу с количеством выявленных преступников по признакам потерпевших. Теперь добавим в таблицу наблюдение, со значением количества потерпевших мужчин старше 18

```python
man_victim_cnt = (victim_feature_grouped['total'][1] - 
                  victim_feature_grouped['total'][2] - 
                  victim_feature_grouped['total'][4])

man_victims = pd.DataFrame(data={'victim_feature': ['man'], 'total': [man_victim_cnt]})
victim_feature_grouped = pd.concat([victim_feature_grouped, man_victims])
victim_feature_grouped
```

Посчитаем доли каждого признака, от общего количества потерпевших

```python
victim_feature_grouped['ratio'] = (
    victim_feature_grouped['total'].div(victim_feature_grouped['total'].values[0])
)
victim_feature_grouped
```

Выводы:
- В период с 2015-2021 годы статус потерпевших получили 5 830 150 лиц (в том числе и юридических);
- 85% потерпевших - физические лица, 15% юридические лица (организации)
- Доля потерпевших женщин - 40%;
- Доля несовершеннолетних - 6%;
- 1.5% потерпевших погибли в результате противоправных действий, а 2.4% получили тяжкий вред здоровью

### Половозрастной портрет преступника и потерпевшего в России

Построим графики с половозрастными признаками преступников и потерпевших (для физических лиц)

```python
fig, ax = plt.subplots(1, 2, figsize=(15,5))

ax[0].set_title('Половозрастные признаки преступности в России')
ax[0].set_ylabel('Доля')
sns.barplot(
    y='ratio', 
    x='sex_age_feature', 
    data=(offender_feature_grouped.query('sex_age_feature.isin(["man", "woman", "juvenile"])')
                                  .sort_values(by='total', ascending=False)), 
    ax=ax[0]
)
ax[0].bar_label(ax[0].containers[0], fmt='%.3f')

ax[1].set_title('Половозрастные признаки потерпевших (физических лиц) в России')
ax[1].set_ylabel('Доля')
sns.barplot(
    y='ratio', 
    x='victim_feature', 
    data=(victim_feature_grouped.query('victim_feature.isin(["man", "woman", "juvenile"])')
                                .sort_values(by='total', ascending=False)), 
    ax=ax[1],
)
ax[1].bar_label(ax[1].containers[0], fmt='%.3f')


plt.show()
```

- Подавляющее большинство преступников - это мужчины старше 18 лет (около 80% всех выявленных преступников)
- Потерпевшие по полу распределены достаточно сбалансировано - 40% женщин 45% мужчин
- Несовершеннолетних преступников - 4% против 6.5% несовершеннолетних потерпевших

### Посмотрим динамику по годам, изменения признаков преступников и потерпевших по России

```python
common_gender_age_crimes_df = gender_age_crimes_df.query('region == "Всего по России "')
common_victims_df = victims_df.query('region == "Всего по России "')
```

```python
offender_feauture_by_years = common_gender_age_crimes_df.pivot_table(
    index='year', 
    columns='sex_age_feature', 
    values='total'
)

offender_feauture_by_years = offender_feauture_by_years.div(offender_feauture_by_years['all'], axis=0)

offender_feauture_by_years
```

Рассчитали доли признаков преступников, по России, по годам

## Проверка гипотез

### Провеим гипотезу об отличии долей женской преступности в Чукотском АО от доли женской преступности по всей России

Так как наличие статистически значимого различия в доли несовершеннолетних преступников в Забайкальском крае от доли по всей России не вызывает сомнений (различие на 46%), то проверим только гипотезу о разности долей женской преступности в Чукотском АО и доли женской преступности по всей России

Сформулируем гипотезы:

$H_{0}: p_1 = p_2\\$
$H_{1}: p_1 \neq p_2$

**Основная (нулевая) гипотеза** - Доля выявленных преступников женщин в Чукотском АО не имеет статистически значимой разницы с долей выявленных преступников женщин в общем по России.

**Альтернативная гипотеза** - Доля выявленных преступников женщин в Чукотском АО имеет статистически значимую разницу с долей выявленных преступников женщин в общем по России.

Напишем функцию для определения вероятности отклонения двух долей

```python
def z_test(p1, p2, successes, trials):
    p_combined = (successes[0] + successes[1]) / (trials[0] + trials[1])
    difference = p1 - p2
    z_value = difference / np.sqrt(
        p_combined * (1 - p_combined) * (1/trials[0] + 1/trials[1]))
    distr = st.norm(0, 1)
    p_value = (1 - distr.cdf(abs(z_value))) * 2
    return p_value
```

Установим уровень значимости в 5% и проведем тест

```python
alpha = .05

russia_values = gender_age_features_by_region.query('index == "Всего по России "').values[0]
region_values = gender_age_features_by_region.query('index == "Чукотский АО"').values[0]

trials1, successes1, p1 = russia_values[0], russia_values[2], russia_values[3]
trials2, successes2, p2 = region_values[0], region_values[2], region_values[3]

p_value = z_test(p1, p2, [successes1, successes2], [trials1, trials2])

print('Доля для России', round(successes1 / trials1, 2))
print('Доля для региона', round(successes2 / trials2, 2))
print()
print('p-значение: ', p_value)
if p_value < alpha:
    print('Отвергаем нулевую гипотезу: между долями есть значимая разница')
else:
    print(
        'Не получилось отвергнуть нулевую гипотезу, '
        'нет оснований считать, что есть значимая разница.'
    )
```

Вывод: Доля выявленных преступников женщин в Чукотском АО значительно больше доли выявленных преступников женщин в общем по России в период с 2015 по 2021 годы

### Проверим гипотезу об отличии долей женской преступности Северо Кавказского федерального округа от доли женской преступности по всей России

Сформулируем гипотезы:

$H_{0}: p_1 = p_2\\$
$H_{1}: p_1 \neq p_2$

**Основная (нулевая) гипотеза** - Доля выявленных преступников женщин в Северо-Кавказском федеральном округе не имеет статистически значимой разницы с долей выявленных преступников женщин в общем по России.

```python
alpha = .05

russia_values = gender_age_features_by_region.query('index == "Всего по России "').values[0]
region_values = gender_age_features_by_region.query('index == "Северо-Кавказский фед.округ"').values[0]

trials1, successes1, p1 = russia_values[0], russia_values[2], russia_values[3]
trials2, successes2, p2 = region_values[0], region_values[2], region_values[3]

p_value = z_test(p1, p2, [successes1, successes2], [trials1, trials2])

print('Доля для России', round(successes1 / trials1, 2))
print('Доля для региона', round(successes2 / trials2, 2))
print()
print('p-значение: ', p_value)

if p_value < alpha:
    print('Отвергаем нулевую гипотезу: между долями есть значимая разница')
else:
    print(
        'Не получилось отвергнуть нулевую гипотезу, '
        'нет оснований считать, что есть значимая разница.'
    )
```
