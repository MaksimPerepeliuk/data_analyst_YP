## Описание исследования

В мобильном приложении "Ненужные вещи" пользователи продают свои ненужные вещи, размещая их на доске объявлений при этом пользователи желающие купить какие-либо вещи, совершают различные события, логи которых и представлены в качестве данных.

## Задачи проекта
Выделить группы пользователей на основе различных метрик, которые склонны часто возвращаться в мобильное приложение, для этого нам нужно ответить на следующие вопросы:

1. Какие пользователи склонны часто возвращаться в мобильное приложение?
2. Какие пользователи часто совершают целевое событие `contacts_show`?

## Исходные данные

В датасете содержатся данные пользователей, впервые совершивших действия в приложении после 7 октября 2019 года.

Датасет **mobile_dataset.csv** содержит колонки:
- `event.time` — время совершения
- `event.name` — название события
- `user.id` — идентификатор пользователя

Расшифровки событий `event.name`:
- `advert_open` — открытие карточки объявления
- `photos_show` — просмотр фотографий в объявлении
- `tips_show` — пользователь увидел рекомендованные объявления
- `tips_click` — пользователь кликнул по рекомендованному объявлению
- `contacts_show` и `show_contacts` — пользователь нажал на кнопку "посмотреть номер телефона" на карточке объявления
- `contacts_call` — пользователь позвонил по номеру телефона на карточке объявления
- `map` — пользователь открыл карту размещенных объявлений
- `search_1` — `search_7` — разные события, связанные с поиском по сайту
- `favorites_add` — добавление объявления в избранное

Датасет **mobile_sources.csv** содержит колонки:
- `userId` — идентификатор пользователя
- `source` — источник, с которого пользователь установил приложение

## Рассчитаем новые признаки для дальнейшего анализа

### Отделим события от сессий

Так как в наших данных отсутсвует информация о конце сессии, рассчитаем разницу в секундах, между соседними событиями каждого пользователя

```python
# сортируем данные по пользователям и времени события
data = data.sort_values(by=['user_id', 'event_time'])
# находим временную разницу между событиями в секундах
data['events_time_diff'] = data.groupby('user_id')['event_time'].diff().dt.seconds
# заполняем образовавшиеся пропуски
data['events_time_diff'] = data['events_time_diff'].fillna(0)
data.head()
```

Для определения сессий возьмем временной промежуток в 30 минут (эмпирическое наблюдение), при котором два соседних события будут считаться событиями двух разных сессий

```python
sessions_delay = 1800
data['new_session'] = data['events_time_diff'].apply(
    lambda diff: 1 if diff >= sessions_delay else 0
)
```

### Расчет новых признаков

Для каждого пользователя посчитаем общую среднюю продолжительность сессии, среднее количество и среднюю продолжительность сессий в день

### Подбор гиперпараметра

Для подбора оптимального количества кластеров используем метрику _silhouette score_

```python
n_clusters = range(2, 30) # объявили диапазон для подбора количества кластеров
scores = []
for i in tqdm(n_clusters):
    # на каждой итерации обучаем модель с новым параметром n_clusters
    km = KMeans(n_clusters=i, random_state=0)
    labels = km.fit_predict(X_sc)
    # рассчитываем метрику silhouette score и сохраняем значение в список scores
    scores.append(silhouette_score(X_sc, labels))
    
# строим линейный график и отмечаем максимальное значение
max_score_clusters = n_clusters[np.argmax(scores)]
plt.figure(figsize=(10, 5))
sns.lineplot(x=n_clusters, y=scores)
plt.axvline(x=max_score_clusters, linestyle='--', color='red')
plt.text(x=max_score_clusters + .3, y=np.min(scores), s='Max',rotation=90)
plt.xticks(n_clusters)
plt.title('Изменение метрики Silhouette Score в зависимости от n_clusters')
plt.xlabel('Количество кластеров n_clusters')
plt.ylabel('Silhuette Score')
plt.show()
```

В данном случае максимум Silhouette Score достигается для n_clusters = 7

## Когортный анализ

### Retention Rate - удержание пользователей

Рассчитаем retention rate, в качестве лайфтайма у нас будут выступать дни, в качестве когорт - метки кластеров `label`. Для начала сформируем данные которые нам потребуются для рассчетов

```python
# получаем источник привлечения для каждого пользователя
users_sources = data.groupby('user_id', as_index=False).agg({'source': 'first'})
# получаем метки кластеров для каждого пользователя
users_sources['label'] = labels

users = data[['user_id', 'event_time', 'date']]
users['date'] = pd.to_datetime(users['date'])
# получаем признак - первая временная метка для каждого пользователя
users = users.merge(
    users.groupby('user_id', as_index=False)
          .agg({'date': 'min'})
          .rename(columns={'date': 'first_ts'}),
    on='user_id',
    how='left'
)
# соединяем таблицы
users = users.merge(users_sources[['user_id', 'label']], on='user_id', how='left')
users.head()
```

Напишем функцию для рассчета Retention Rate

```python
def get_retention(
    profiles,
    observation_date,
    horizon_days,
    dimensions=[],
    ignore_horizon=False,
):

    # исключаем пользователей, не «доживших» до горизонта анализа
    last_suitable_acquisition_date = observation_date
    if not ignore_horizon:
        last_suitable_acquisition_date = observation_date - timedelta(
            days=horizon_days - 1
        )
    result_raw = profiles.query('date <= @last_suitable_acquisition_date')

    # добавляем признак лайфтайм для каждого пользователя
    result_raw['lifetime'] = (
        profiles['date'] - result_raw['first_ts']
    ).dt.days

    # рассчитываем удержание
    result_grouped = result_raw.pivot_table(
        index=dimensions, # индексы - когорты
        columns='lifetime',
        values='user_id',
        aggfunc='nunique',
    )
    
    # выделяем значения размеров когорт
    cohort_sizes = (
        result_raw.groupby(dimensions) 
        .agg({'user_id': 'nunique'})
        .rename(columns={'user_id': 'cohort_size'})
    )
    
    # присоединяем размеры когорт к таблице
    result_grouped = cohort_sizes.merge(
        result_grouped, on=dimensions, how='left'
    ).fillna(0)
    
    # делим каждую строку на размеры когорт
    result_grouped = result_grouped.div(result_grouped['cohort_size'], axis=0)

    # восстанавливаем столбец с размерами когорт
    result_grouped['cohort_size'] = cohort_sizes

    # возвращаем таблицу удержания и сырые данные
    return result_raw, result_grouped
```

Рассчитаем Retention Rate и построим тепловую карту

```python
horizon_days = 28 # горизонт анализа - максимальный лайфтайм
observation_date = data['date'].max() # момент анализа - максимальная дата

retention_raw, retention = get_retention(
    users, observation_date, horizon_days, dimensions=['label'], ignore_horizon=True
)  # кортами у нас будут номера кластеров - label

users['cohort'] = 'all_users' # добавляем общую когорту

# получаем конверсию для общей когорты всех пользователей на каждый лайфтайм
all_users_retention_raw, all_users_retention = get_retention(
    users, observation_date, horizon_days, dimensions=['cohort'], ignore_horizon=True
)

# соединяем таблицы
retention = pd.concat([retention, all_users_retention], axis=0)

display(retention[['cohort_size']])

# строим тепловую карту
plt.figure(figsize=(20, 8))
sns.heatmap(retention.drop(columns=['cohort_size', 0]), annot=True, fmt='.2%')
plt.title('Тепловая карта удержания')
plt.xlabel('Лайфтайм')
plt.ylabel('Кластеры')
plt.tight_layout()
plt.show()
```

Первые 3 когорты (кластеры 0, 1, 2) по показателю Retention Rate гораздо хуже остальных когорт, в том числе и когорты куда мы поместили всех пользователей `all_users` для наглядности

Построим линейный график с кривыми изменения Retention Rate когорт

```python
report = retention.drop(columns = ['cohort_size', 0])

report.T.plot(
    grid=True,  # добавляем сетку
    xticks=list(report.columns.values),  # отметки на оси X — названия колонок
    figsize=(15, 5),  # размер графика
)
plt.xlabel('Лайфтайм')
plt.title('Кривые удержания по кластерам')
plt.show()
```

Выводы по графикам:
- Кластеры под номерами 0, 1, 2 показали гораздо худший процент удержания чем все остальные когорты
- Кластеры под номерами 3, 4, 6 показали существенно лучший процент удержания, чем общий Retention для всех пользователей (когорта `all_users`)

### Conversion Rate - коэффициент конверсии в целевое действие `contacts show`

По аналогии с Retention Rate проведем анализ конверсии. Для начала сформируем нужные нам данные

```python
users = data[['user_id', 'event_name', 'event_time', 'date']]

contacts_show_first_ts = (
    users.query('event_name == "contacts_show"')
            .groupby('user_id', as_index=False)
            .agg({'event_time': 'min'})
            .rename(columns={'event_time': 'first_contacts_show_ts'})
)

users = users.merge(contacts_show_first_ts, on='user_id', how='left')
users = users.merge(users_sources[['user_id', 'label']], on='user_id', how='left')
users = users.merge(
    users.groupby('user_id', as_index=False)
          .agg({'date': 'min'})
          .rename(columns={'date': 'first_ts'}),
    on='user_id',
    how='left'
)
users['first_ts'] = pd.to_datetime(users['first_ts'])

users.head()
```

```python
n_users = (users.groupby('user_id', as_index=False)
                 .agg({'first_contacts_show_ts': 'last'})['first_contacts_show_ts']
                 .isna().sum())

print(
     'Количество пользователей не '
     'совершивших ни одного просмотра контактов' 
     f' = {n_users} или {round(n_users / len(users["user_id"].unique()) * 100)}%'
)
```

Из-за того, что не все пользователи впринципе делали событие `contacts_show` у нас образовались пропуски. Для корректного расчета признака `lifetime` и CR пропуски надо удалить, сделаем это внутри функции для расчета конверсии, после расчета размера когорт

```python
def get_conversion(
    profiles,
    observation_date,
    horizon_days,
    dimensions=[],
    ignore_horizon=False,
):
    # вычисляем размеры когорт
    cohort_sizes = (
        profiles.groupby(dimensions)
                .agg({'user_id': 'nunique'})
                .rename(columns={'user_id': 'cohort_size'})
    )
    
    # удаляем пользователей не совершивших ни одного просмотра контактов
    profiles = profiles.dropna()
    
    # исключаем пользователей, не «доживших» до горизонта анализа
    last_suitable_acquisition_date = observation_date
    if not ignore_horizon:
        last_suitable_acquisition_date = observation_date - timedelta(
            days=horizon_days - 1
        )
    result_raw = profiles.query('date <= @last_suitable_acquisition_date')

    # рассчитываем лайфтайм для каждого просмотра контактов
    result_raw['lifetime'] = (
        result_raw['first_contacts_show_ts'] - result_raw['first_ts']
    ).dt.days
    
    # строим «треугольную» таблицу конверсии
    result = result_raw.pivot_table(
        index=dimensions, columns='lifetime', values='user_id', aggfunc='nunique'
    )
    # считаем сумму с накоплением для каждой строки
    result = result.fillna(0).cumsum(axis = 1)
    # добавляем размеры когорт в таблицу конверсии
    result = cohort_sizes.merge(result, on=dimensions, how='left').fillna(0)
    # делим каждую «ячейку» в строке на размер когорты
    # и получаем conversion rate
    result = result.div(result['cohort_size'], axis=0)
    # восстанавливаем размеры когорт
    result['cohort_size'] = cohort_sizes

    # возвращаем таблицу конверсий и сырые данные
    return result_raw, result
```

Рассчитаем Conversion Rate и строим графики

```python
horizon_days = 28 # горизонт анализа - максимальный лайфтайм
observation_date = data['date'].max() # момент анализа - максимальная дата

# получаем конверсию по когортам на каждый лайфтайм
conversion_raw, conversion = get_conversion(
    users, observation_date, horizon_days, dimensions=['label'], ignore_horizon=True
)

users['cohort'] = 'all_users' # добавляем общую когорту для всех пользователей

# получаем общую конверсию для всех пользователей на каждый лайфтайм
all_users_conversion_raw, all_users_conversion = get_conversion(
    users, observation_date, horizon_days, dimensions=['cohort'], ignore_horizon=True
)

# соединяем таблицы
conversion = pd.concat([conversion, all_users_conversion], axis=0)

display(conversion[['cohort_size']])

# строим тепловую карту
plt.figure(figsize=(20, 6))
sns.heatmap(conversion.drop(columns=['cohort_size']), annot=True, fmt='.2%')
plt.title('Тепловая карта конверсии')
plt.xlabel('Лайфтайм')
plt.ylabel('Кластеры')
plt.show()

# строим линейный график
report = conversion.drop(columns = ['cohort_size', 0])

report.T.plot(
    grid=True,  # добавляем сетку
    xticks=list(report.columns.values),  # отметки на оси X — названия колонок
    figsize=(15, 5),  # размер графика
)
plt.xlabel('Лайфтайм')  # название оси X
plt.title('Кривые конверсии по кластерам')  # название графика
plt.show()
```

Выводы по графикам:
- Лучше всего по проценту конверсии показывает себя когорта под номером 4
- Кластеры под номерами 0, 1, 2 показали гораздо худший процент конверсии чем все остальные когорты
- Кластеры под номерами 3, 4, 6 показали существенно лучший процент конверсии, чем общий CR для всех пользователей (когорта `all_users`)
