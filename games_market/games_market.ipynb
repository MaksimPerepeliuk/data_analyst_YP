# Исследование рынка компьютерных и консольных игр

## Описание проекта
Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation)

## Цель проекта

- Нам нужно выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.

## Описание данных
- `Name` — название игры
- `Platform` — платформа
- `Year_of_Release` — год выпуска
- `Genre` — жанр игры
- `NA_sales` — продажи в Северной Америке (миллионы проданных копий)
- `EU_sales` — продажи в Европе (миллионы проданных копий)
- `JP_sales` — продажи в Японии (миллионы проданных копий)
- `Other_sales` — продажи в других странах (миллионы проданных копий)
- `Critic_Score` — оценка критиков (максимум 100)
- `User_Score` — оценка пользователей (максимум 10)
- `Rating` — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.

## Анализ данных

### Посчитаем суммарные продажи во всех регионах

```python
data['total_sales'] = data['na_sales'] + data['eu_sales'] + data['jp_sales'] + data['other_sales']
```

```python
data.head()
```

### Посмотрим, на количество выпущенных игр по годам:

```python
(data.groupby('year_of_release')
    .agg({'name': ['count']})
    .plot(kind='bar', figsize=(9,7), legend=False))

plt.title('Количество выпущенных игр по годам')
plt.xlabel('Год выпуска')
plt.ylabel('Количество игр')
plt.show()
```

На графике виден резкий скачок выпущенных игр после 2001 года. Пик выпуска игр приходится на 2008 и 2009 годы, далее идет довольно резкое падение.

```python
round(len(data[data['year_of_release'] < 2001]) / len(data) * 100, 3)
```

В имеющихся данных доля игр выпущенных до 2001 года составлеет 14%

### Посмотрим на суммарные продажи каждой платформы по годам

```python
sales_by_platform = (data.groupby('platform')
     .agg({'total_sales': ['sum']})
     .sort_values(by=('total_sales', 'sum'), ascending=False))

sales_by_platform.plot(
    kind='bar', 
    figsize=(9, 7), 
    legend=False, 
    title='Количество проданных копий по платформам', 
)
plt.xlabel('Платформы')
plt.ylabel('Копий (млн)')
plt.show()
```

На графике четко видны 6 платформ лидеров по количеству проданных копий

### Посмотрим на распределение выпущенных игр по годам для каждой из 6 платформ - лидеров по продажам

Для этого построим столбчатые графики

```python
most_sales_platform = np.array_split(sales_by_platform.index[:6], 2)
```

```python
for i in range(len(most_sales_platform)):
    for j in range(len(most_sales_platform[i])):
        platform = most_sales_platform[i][j]
        platform_df = data[data['platform'] == platform]
        (platform_df.groupby('year_of_release')
                    .agg({'name': 'count'})
                    .plot(kind='bar', legend=False, ax=plt.subplot(2, 3, j+1), figsize=(30, 15)))
        plt.title(f'Количество выпущенных игр на {platform} по годам', {'fontsize': 18})
        plt.xlabel('Год выпуска')
        plt.ylabel('Количество игр')
        plt.tick_params(labelsize=16)
    plt.show()
```

По графикам видно, что "время жизни" платформы около 10 лет, пик выпуска игр достигается примерно к середине этого периода (5 лет)

### Определим актуальные платформы

Для этого сделаем срез - из последнего года имеющегося у нас периода вычтем половину времени жизни платформы т.е. 2016 - 5 лет = 2011 год

```python
data = data[data['year_of_release'] >= 2011]
```

### Посмотрим на популярность платформ по количеству проданных копий

```python
total_sales_grouped = (
    data.groupby('platform')
        .agg({'total_sales': ['sum']})
        .sort_values(
            by=('total_sales', 'sum'),
            ascending=False
        )
)

total_sales_grouped.plot(
    kind='bar', 
    legend=False,
    title='Количество проданных копий по платформам с 2011 года',
    figsize=(9, 7)
)

plt.ylabel('Копий (млн)')
plt.xlabel('Платформы')
plt.show()
```

### Посмотрим на количество проданных копий по годам для 6 самых популярных платформ

```python
most_popular_platforms = total_sales_grouped.index[:6]
most_popular_platforms_chunked = np.array_split(most_popular_platforms, 2)
for i in range(len(most_popular_platforms_chunked)):
    for j in range(len(most_popular_platforms_chunked[i])):
        platform = most_popular_platforms_chunked[i][j]
        platform_df = data[data['platform'] == platform]
        (platform_df.groupby('year_of_release')
                    .agg({'total_sales': 'sum'})
                    .plot(kind='bar', legend=False, ax=plt.subplot(2, 3, j+1), figsize=(30, 15)))
        plt.title(f'Количество проданных копий {platform} по годам', {'fontsize': 18})
        plt.xlabel('Год выпуска')
        plt.ylabel('Копий (млн)')
        plt.tick_params(labelsize=16)
    plt.show()
```

- Платформы, выпуск игр которых начинается с 2013 года (PS4, XOne), на 2016 год увеличили продажи, хотя и прошли через максимум
- Остальные платформы, на 2016 год, упали по продажам и имеют отрицательный тренд

### Посмотрим на распределение количества проданных копий игр по платформам

```python
most_popular_platforms_df = data[data['platform'].isin(most_popular_platforms)]
(most_popular_platforms_df.boxplot(
          figsize=(13,7),
          column='total_sales',
          by='platform'
      )
)
plt.title('Распределение значений проданных копий по популярным платформам')
plt.ylabel('Копий (млн)')
plt.xlabel('Платформы')
plt.show()
```

Слишком много выбросов (хвосты распределений) мешают увидеть ящик, поэтому масштабируем график

```python
most_popular_platforms_df = data[data['platform'].isin(most_popular_platforms)]
(most_popular_platforms_df.boxplot(
          figsize=(13,7),
          column='total_sales',
          by='platform'
      )
)
plt.title('Распределение значений проданных копий по популярным платформам')
plt.ylabel('Копий (млн)')
plt.xlabel('Платформы')
plt.ylim(top=2, bottom=0)
plt.show()
```

Выводы по графику:
- по всем платформам наблюдается большое количество выбросов
- выбросы которые видны на графике это - популярные игры которые продавались намного большим суммарным тиражом, чем средние (медианные) продажи по платформе, это абсолютно соответствует логике распределения любых продаж (небольшое количество товаров пользуется повышенным спросом)
- выбросы составляют довольно весомый вклад в продажи игр по всем платформам

### Исследуем связь продаж с некоторыми признаками

#### Посмотрим на корреляцию продаж и оценок

## Проверим гипотезы

Для проверки гипотез следует очистить данные от пропусков т.к. любая математическая операция с NaN будет возвращать NaN. Также удалим наблюдения с пользовательской оценкой tbd так как мы выяснили ранее, что этот рейтинг по факту означает отсутствие рейтинга

```python
data = data.query('user_score != "tbd"').dropna()
data['user_score'] = data['user_score'].astype('float')
```

### 1. Средние пользовательские оценки платформ Xbox One и PC одинаковы

$H_{0}: \mu_{XOne} = \mu_{pc}\\$
$H_{1}: \mu_{XOne} \neq \mu_{pc}$

**Основная (нулевая) гипотеза** - средние пользовательские рейтинги платформ Xbox One и PC не имеют статистически значимых различий;

**Альтернативная гипотеза** - средние пользовательские рейтинги платформ Xbox One и PC имеют статистически значимые различия.

Проверим двустороннюю гипотезу о равенстве средних двух генеральных совокупностей для независимых выборок при уровне значимости в 5%:

```python
alpha = 0.05

xone_user_scores = data[data['platform'] == 'XOne']['user_score']
pc_user_scores = data[data['platform'] == 'PC']['user_score']

results = st.ttest_ind(
    xone_user_scores,
    pc_user_scores,
)

print(xone_user_scores.mean().round(3), pc_user_scores.mean().round(3))
print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу') 
```

Вывод: у нас нет оснований полагать, что пользовательские оценки 'XOne' отличаются от пользовательских оценок 'PC'

### 2. Средние пользовательские рейтинги жанров Action и Sports разные.

$H_{0}: \mu_{action} = \mu_{sports}\\$
$H_{1}: \mu_{action} \neq \mu_{sports}$

**Основная (нулевая) гипотеза** - средние пользовательские оценки жанров action и sports не имеют статистически значимых различий;

**Альтернативная гипотеза** - средние пользовательские оценки жанров action и sports имеют статистически значимые различия.

Проверим двустороннюю гипотезу о равенстве средних двух генеральных совокупностей для независимых выборок при уровне значимости в 5%:

```python
alpha = 0.05

action_user_scores = data[data['genre'] == 'Action']['user_score']
sports_user_scores = data[data['genre'] == 'Sports']['user_score']

results = st.ttest_ind(
    action_user_scores,
    sports_user_scores,
)


print(action_user_scores.mean().round(3), sports_user_scores.mean().round(3))
print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу') 
```

Вывод: у нас есть основания полагать, что пользовательские оценки у жанра 'Action' в среднем больше, чем у жанра 'Sports'
